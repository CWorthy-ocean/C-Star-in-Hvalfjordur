
## Quantifying $pCO_2$ Bias from Mean-State Errors

### 1. Decomposition of $pCO_2$ Bias

First, we quantify how biases in the modeled mean state (DIC, alkalinity, and temperature) propagate into errors in simulated $pCO_2$.


Using a first-order Taylor expansion, the total bias in $pCO_2$ ($\Delta pCO_2^{total}$) can be approximated as the sum of the linear contributions from each state variable:

$$
\Delta pCO_2^{total} \approx \left( \frac{\partial pCO_2}{\partial DIC} \right) \Delta DIC + \left( \frac{\partial pCO_2}{\partial Alk} \right) \Delta Alk + \left( \frac{\partial pCO_2}{\partial T} \right) \Delta T
$$

Where:
* **$\Delta DIC, \Delta Alk, \Delta T$** are the mean-state biases (Model $-$ Observed).
* **$\frac{\partial pCO_2}{\partial X}$** are the partial derivatives (sensitivities) evaluated at the observed reference state.


We find that the alkalinity and DIC biases in the model largely cancel out, but any temperature bias in the model will propagate into $pCO_2$ space 



```python
import numpy as np
import PyCO2SYS as pyco2
import pandas as pd

# -------------------------------
# helper: compute pCO2
# -------------------------------
def compute_pCO2(alk_m3, dic_m3, sal, temp, pres):
    rho = 1026.0 
    out = pyco2.sys(
        par1=(alk_m3 / rho) * 1000,
        par2=(dic_m3 / rho) * 1000,
        par1_type=1, par2_type=2,
        salinity=sal, temperature=temp, pressure=pres,
    )
    return out["pCO2"]

# -------------------------------
# derivatives
# -------------------------------
def get_derivatives(state, eps=30, eps_t=0.1):
    alk, dic = state["alk"], state["dic"]
    s, t, p = state["sal"], state["temp"], state["pres"]
    
    # Chemistry First derivatives
    dpc_dDIC = (compute_pCO2(alk, dic + eps, s, t, p) - 
                compute_pCO2(alk, dic - eps, s, t, p)) / (2 * eps)
    
    dpc_dALK = (compute_pCO2(alk + eps, dic, s, t, p) - 
                compute_pCO2(alk - eps, dic, s, t, p)) / (2 * eps)
    
    # Temperature First derivative
    dpc_dT = (compute_pCO2(alk, dic, s, t + eps_t, p) - 
              compute_pCO2(alk, dic, s, t - eps_t, p)) / (2 * eps_t)
    
    return dpc_dDIC, dpc_dALK, dpc_dT

# -------------------------------
# main calculation
# -------------------------------
def dpCO2_bias_analysis(obs, model):
    # 1. Calculate Sensitivities at Observed State
    dDIC_sens, dALK_sens, dTemp_sens = get_derivatives(obs)

    # 2. Mean state biases
    delta_DIC = model["dic"] - obs["dic"]
    delta_ALK = model["alk"] - obs["alk"]
    delta_Temp = model["temp"] - obs["temp"]

    # 3. Individual components
    error_DIC = dDIC_sens * delta_DIC
    error_ALK = dALK_sens * delta_ALK
    error_Temp = dTemp_sens * delta_Temp
    
    # 4. Actual total for reference
    actual_total = compute_pCO2(model['alk'], model['dic'], model['sal'], model['temp'], model['pres']) - \
                   compute_pCO2(obs['alk'], obs['dic'], obs['sal'], obs['temp'], obs['pres'])

    # Summary table
    data = {
        "Component": [
            "DIC First-Order", 
            "ALK First-Order", 
            "Temp First-Order", 
            "Total Approximation", 
            "Actual Bias"
        ],
        "Sensitivity": [
            dDIC_sens, 
            dALK_sens, 
            dTemp_sens, 
            np.nan, 
            np.nan
        ],
        "Bias (ΔX)": [
            delta_DIC, 
            delta_ALK, 
            delta_Temp, 
            np.nan, 
            np.nan
        ],
        "pCO2 Error (uatm)": [
            error_DIC, 
            error_ALK, 
            error_Temp,
            error_DIC + error_ALK + error_Temp,
            actual_total
        ]
    }
    
    return pd.DataFrame(data)

# -------------------------------
# example usage
# -------------------------------
model = dict(alk=2320, dic=2110.0, sal=33.75, temp=11.5, pres=0)
obs = dict(alk=2380, dic=2160.0, sal=33.75, temp=10, pres=0)

df_results = dpCO2_bias_analysis(obs, model)
df_results
```

### 2. Intervention Anomaly Error

When evaluating an alkalinity intervention of magnitude $I$, the error in the predicted $pCO_2$ drawdown (the anomaly) is driven by the difference in sensitivities between the modeled and observed states. We can estimate this by taking the difference of the model and observed sensitivities of pCO2 to Alk:

$$
\text{Error}_{\text{anomaly}} \approx \left[ \left( \frac{\partial pCO_2}{\partial Alk} \right)_{\text{model}} - \left( \frac{\partial pCO_2}{\partial Alk} \right)_{\text{obs}} \right] \times I
$$

And compare with the error we get by substracting the full CO2SYS calculation of the modeled and observed states:

$$
\text{Error}_{\text{actual}} = \left( f(Alk_m + I, DIC_m, ...) - f(Alk_m, DIC_m, ...) \right) - \left( f(Alk_o + I, DIC_o, ...) - f(Alk_o, DIC_o, ...) \right)
$$

**We find that the error in pCO2 when the added alkalinity has a concentration of 45 mmol/m$^3$ is about 10%**

```python
import numpy as np
import PyCO2SYS as pyco2
import pandas as pd

# -------------------------------
# helper: compute pCO2
# -------------------------------
def compute_pCO2(alk_m3, dic_m3, sal, temp, pres):
    rho = 1026.0 
    out = pyco2.sys(
        par1=(alk_m3 / rho) * 1000,
        par2=(dic_m3 / rho) * 1000,
        par1_type=1, par2_type=2,
        salinity=sal, temperature=temp, pressure=pres,
    )
    return out["pCO2"]

# -------------------------------
# derivative helper
# -------------------------------
def get_alk_sensitivity(alk, dic, sal, temp, pres, eps=30):
    p_plus  = compute_pCO2(alk + eps, dic, sal, temp, pres)
    p_minus = compute_pCO2(alk - eps, dic, sal, temp, pres)
    return (p_plus - p_minus) / (2 * eps)

# -------------------------------
# main calculation
# -------------------------------
def quantify_intervention_error(obs, model, intervention_val=45):
    # 1. Sensitivities
    sens_obs   = get_alk_sensitivity(**obs)
    sens_model = get_alk_sensitivity(**model)
    
    # 2. Taylor Estimates for Drawdown
    taylor_drawdown_obs   = sens_obs * intervention_val
    taylor_drawdown_model = sens_model * intervention_val

    # 3. Actual Full CO2SYS Anomalies
    actual_drawdown_obs = compute_pCO2(obs['alk'] + intervention_val, obs['dic'], 
                                       obs['sal'], obs['temp'], obs['pres']) - \
                          compute_pCO2(obs['alk'], obs['dic'], 
                                       obs['sal'], obs['temp'], obs['pres'])
    
    actual_drawdown_model = compute_pCO2(model['alk'] + intervention_val, model['dic'], 
                                         model['sal'], model['temp'], model['pres']) - \
                            compute_pCO2(model['alk'], model['dic'], 
                                         model['sal'], model['temp'], model['pres'])
    
    actual_anomaly_error = actual_drawdown_model - actual_drawdown_obs

    # 4. Taylor Expansion Estimate of the Anomaly Error
    taylor_anomaly_error = (sens_model - sens_obs) * intervention_val

    # Summary table
    data = {
        "Metric": [
            "ALK Sensitivity (Obs)",
            "ALK Sensitivity (Model)",
            "Taylor Predicted pCO2 anomaly(Obs)",
            "Taylor Predicted pCO2 anomaly (Model)",
            "Actual pCO2 anomaly (Obs State)", 
            "Actual pCO2 anomaly (Model State)", 
            "Actual error", 
            "Error estimate"
        ],
        "Value": [
            sens_obs,
            sens_model,
            taylor_drawdown_obs,
            taylor_drawdown_model,
            actual_drawdown_obs, 
            actual_drawdown_model, 
            actual_anomaly_error,
            taylor_anomaly_error
        ],
        "Units": [
            "uatm/(mmol/m3)", "uatm/(mmol/m3)", 
            "uatm", "uatm", "uatm", "uatm", "uatm", "uatm"
        ]
    }
    
    return pd.DataFrame(data)

# -------------------------------
# example usage
# -------------------------------
obs   = dict(alk=2380, dic=2160.0, sal=33.75, temp=10.0, pres=0)
model = dict(alk=2320, dic=2110.0, sal=33.75, temp=11.5, pres=0)

df_intervention = quantify_intervention_error(obs, model, intervention_val=45)
df_intervention
```
